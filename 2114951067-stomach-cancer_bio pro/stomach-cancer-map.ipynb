```python
#import pandas

import pandas as pd
```

```python
#read in csv

cancer = pd.read_csv('2019.csv')
```

```python
stdz = cancer[cancer['age_name'] == 'Age-standardized']
stdz
```

```python
stdz.to_csv('age-standardized.csv', index=False)
```

```python
okla = stdz[stdz['location_name'].str.contains('Oklahoma', na=False)]
okla
```

```python
okla.to_csv('oklahoma.csv', index=False)
```

```python
okla_AIAN = okla[okla['race_name'] == 'AIAN']
okla_AIAN
```

```python
okla_AIAN.to_csv('okla_AIAN.csv', index=False)
```
