```python
import pandas as pd
```

```python
y2019 = pd.read_csv('2019.csv')
y2018 = pd.read_csv('2018.csv')
y2017 = pd.read_csv('2017.csv')
y2016 = pd.read_csv('2016.csv')
y2015 = pd.read_csv('2015.csv')
y2014 = pd.read_csv('2014.csv')
y2013 = pd.read_csv('2013.csv')
y2012 = pd.read_csv('2012.csv')
y2011 = pd.read_csv('2011.csv')
y2010 = pd.read_csv('2010.csv')
y2009 = pd.read_csv('2009.csv')
y2008 = pd.read_csv('2008.csv')
y2007 = pd.read_csv('2007.csv')
y2006 = pd.read_csv('2006.csv')
y2005 = pd.read_csv('2005.csv')
y2004 = pd.read_csv('2004.csv')
y2003 = pd.read_csv('2003.csv')
y2002 = pd.read_csv('2002.csv')
y2001 = pd.read_csv('2001.csv')
y2000 = pd.read_csv('2000.csv')
```

```python
all_years = [y2019,
y2018,
y2017,
y2016,
y2015,
y2014,
y2013,
y2012,
y2011,
y2010,
y2009,
y2008,
y2007,
y2006,
y2005,
y2004,
y2003,
y2002,
y2001,
y2000]
```

```python
combined = pd.concat(all_years, ignore_index=True)
combined
```

```python
stdz = combined[combined['age_name'] == 'Age-standardized']
stdz
```

```python
okla_allyears = stdz[stdz['location_name'] == 'Oklahoma']
okla_allyears
```

```python
okla_allyears.to_csv('oklahoma 2000-2019.csv', index=False)
```

```python
wide_df = okla_allyears.pivot(index='year', columns='race_name', values='val')
wide_df
```

```python
wide_df.to_csv('pivoted oklahoma.csv', index=True)
```
